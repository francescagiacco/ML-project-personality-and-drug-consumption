```python
# function to covert nomi sostanze to a np.array to append to the dataset
# funziona

def colonna_gruppo(nomi_sostanze, data):
    '''
    input:
    nomi_sostanze : 
    
    '''
    z = []
    for sostanza in nomi_sostanze:
         #print(data[sostanza[0]])
         z.append((data[sostanza[0]] * sostanza[1]))
         #print(z)
    #print(depressant[0])
    #print(depressant[1])
    column = np.zeros(len(z[0]))
    for i in range(len(z)):
        column = np.add(column, z[i])
    
    return column
    
column_depressant = colonna_gruppo(depressants, data)
column_stimuants = colonna_gruppo(stimulants, data)
column_hallucinogens = colonna_gruppo(hallucinogens, data)
column_depressants_no_vino = colonna_gruppo(depressants_no_vino, data)

# create a new dataframe to append
print(column_depressants_no_vino)
```
