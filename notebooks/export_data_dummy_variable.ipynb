```python
import pandas as pd
import utils as utl
```

## Create the dictionaries

```python
etnicity = {'asian' : -0.50212, 'black_asian' : 1.90725, 'white_asian' : 0.126001,
             'white_black' : -0.22166, 'other' : 0.11440, 'white' : -0.31685, 
             'other2': -1.10702 } 

country = {'australia' : -0.09765, 'canada' : 0.24923, 'new_zealand' : -0.46841, 'other' : -0.28519,
              'ireland' : 0.21128, 'uk' : 0.96082, 'usa' : -0.57009}

gender = {'male' : -0.48246, 'female' : 0.48246}
```

```python
data = pd.read_csv('drug_consumption.csv')
data.replace(0.12600, 0.126001, inplace = True) #tips tranform to float, zero were rounded 
data.rename(columns={"country ":"country"}, inplace = True)  #BUG BUG BUG country got the space, remove it

data.columns
```

## Main loop where I call the functions

```python
list_dictionary = [etnicity, country, gender]
list_column_name = ['etnicity', 'country', 'gender']
j = 0
for i, column_name in zip(list_dictionary, list_column_name):
    if j == 0:
        dictionary = utl.create_dictionary(data[column_name], i)
        final_df = utl.append_in_dataframe(data, dictionary)
    else:
        dictionary = utl.create_dictionary(final_df[column_name], i)
        final_df = utl.append_in_dataframe(final_df, dictionary)
    j += 1
    final_df.drop(column_name, axis = 1, inplace = True)
    #print(i, column_name)
final_df.head()
```

## Export the dataset

```python
path = '.\data_processed'
file_name = '\data_dummy_variable'

# change a name based on wich column you have changed 
for column in list_column_name:
    file_name += '_' + column
final_path = path + file_name + '.csv'
# example: print(final_path) -> .\data_processed\data_dummy_variable_etnicity_country_gender.csv

final_df.to_csv(final_path ) #This time I won, Windows 1 - 1 Zaz
```
